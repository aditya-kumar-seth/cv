```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image in grayscale mode
image = cv2.imread('input.jpg', cv2.IMREAD_GRAYSCALE)

# Check if image is loaded properly
if image is None:
    raise ValueError('Image not found. Please check the file path.')

# Convert image to float32 for mathematical operations
image_float = image.astype(np.float32)

# Normalize image to range [0, 1]
image_norm = image_float / 255.0

# -------------------------------------------------
# 1. Negative Transformation
# Formula: s = 1 - r
# -------------------------------------------------
negative = 1.0 - image_norm
negative = (negative * 255).astype(np.uint8)

# -------------------------------------------------
# 2. Log Transformation
# Formula: s = c * log(1 + r)
# -------------------------------------------------
log_transformed = np.log1p(image_norm)

# Normalize log output back to [0, 1]
log_transformed = log_transformed / np.max(log_transformed)

# Convert back to uint8
log_transformed = (log_transformed * 255).astype(np.uint8)

# -------------------------------------------------
# 3. Power-Law (Gamma) Transformation
# Formula: s = c * r^gamma
# -------------------------------------------------
gamma = 0.5   # gamma < 1 brightens image
power_law = np.power(image_norm, gamma)

# Normalize output to [0, 1]
power_law = power_law / np.max(power_law)

# Convert back to uint8
power_law = (power_law * 255).astype(np.uint8)

# -------------------------------------------------
# 4. Contrast Stretching
# Formula:
# s = (r - r_min) / (r_max - r_min)
# -------------------------------------------------
r_min = np.min(image_float)
r_max = np.max(image_float)

contrast_stretched = (image_float - r_min) / (r_max - r_min)
contrast_stretched = (contrast_stretched * 255).astype(np.uint8)

# -------------------------------------------------
# 5. Thresholding Example
# -------------------------------------------------
threshold_value = 127
binary_image = np.where(image > threshold_value, 255, 0).astype(np.uint8)

# -------------------------------------------------
# Display Images
# -------------------------------------------------
plt.figure(figsize=(15, 10))

images = [
    ('Original', image),
    ('Negative', negative),
    ('Log Transform', log_transformed),
    ('Gamma Transform', power_law),
    ('Contrast Stretching', contrast_stretched),
    ('Thresholding', binary_image)
]

for i, (title, img) in enumerate(images):
    plt.subplot(2, 3, i + 1)
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
